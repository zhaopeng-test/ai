决策树实战task
1. 基于iris_data.csv 数据，建立决策树模型，评估模型表项
2. 可视化决策树结构
3. 修改min_samples_leaf 参数，对比模型结果


```python
import pandas as pd 
import numpy as np

data = pd.read_csv('iris.csv')
data.head()
```

```python
x = data.drop(['variety'], axis=1)
y = data.loc[:, 'variety']
x.head()
print(x.shape, y.shape)
```

```python
#establish the decision tree model
from sklearn import tree
#信息熵增益最大化
dc_tree = tree.DecisionTreeClassifier(criterion='entropy', min_samples_leaf=5)
dc_tree.fit(x,y)
```

```python
#envaluete the model

y_predict = dc_tree.predict(x)

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y, y_predict)
print(accuracy)
```

```python
#visualize the tree

from matplotlib import pyplot as plt 
fig = plt.figure(figsize=(10,10))
tree.plot_tree(dc_tree)
#tree.plot_tree(dc_tree, filled=np.True_, feature_names=['sepal.length','sepal.width','petal.length','petal.width'], class_names=['setosa','versicolor','virgrinca'])
```

```python
fig2 = plt.figure(figsize=(10,10))
tree.plot_tree(dc_tree)
tree.plot_tree(dc_tree, filled=np.True_, feature_names=['sepal.length','sepal.width','petal.length','petal.width'], class_names=['setosa','versicolor','virgrinca'])
```

```python
#modify min_samples_leaf
dc_tree1 = tree.DecisionTreeClassifier(criterion='entropy', min_samples_leaf=1)
dc_tree1.fit(x,y)
fig3 = plt.figure(figsize=(10,10))
tree.plot_tree(dc_tree1)
tree.plot_tree(dc_tree1, filled=np.True_, feature_names=['sepal.length','sepal.width','petal.length','petal.width'], class_names=['setosa','versicolor','virgrinca'])
```
