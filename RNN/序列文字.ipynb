任务:基于flare文本数据，建立LSTM模型，预测序列文字:
1.完成数据预处理，将文字序列数据转化为可用于LSTM输入的数据
2.查看文字数据预处理后的数据结构，并进行数据分离操作
3.针对字符串输入(” flare is a teacher in ai industry. He obtained his phd in Australia.”)，预测其对应的后续字符备注:模型结构:单层LSTM，输出有20个神经元;每次使用前20个字符预测第21个字符

```python
#load the data
data = open('flare').read()
data = data.replace('\n', ' ').replace('\r', ' ')
print(data)
```

```python
#字符去重处理
letters = list(set(data))
print(letters)
num_letters = len(letters)
print(num_letters)
```

```python
#建立字典
int2char = {i:c for i, c in enumerate(letters)}
print(int2char)
```

```python
char2int = {c:i for i, c in enumerate(letters)}
print(char2int)
```

```python
from tkinter import Y
import numpy as np
from keras.utils import to_categorical

def extract_data(data, slide):
    x = []
    y = []
    for i  in range(0, len(data) - slide):
        x.append([a for a in data[i:i+slide]])
        y.append(data[i+slide])
    return x, y


def char2int_Data(x, y, char_to_int):
    x_int = []
    y_int = []
    for i in range(len(x)):
        x_int.append([char_to_int[cha] for cha in x[i]])
        y_int.append([char_to_int[cha] for cha in y[i]])
    return x_int, y_int

def data_preprocess(data, slide, num_letters, char_to_int):
    char_Data = extract_data(data, slide)
    int_Data = char2int_Data(char_Data[0], char_Data[1], char_to_int)
    input = int_Data[0]
    output = list(np.array(int_Data[1]).flatten())
    input_data_reshaped = np.array(input).reshape(len(input), slide)
    new = np.random.randint(0, 10, size=[input_data_reshaped.shape[0], input_data_reshaped.shape[1], num_letters])
    for i in range(input_data_reshaped.shape[0]):
        for j in range(input_data_reshaped.shape[1]):
            new[i,j,:] = to_categorical(input_data_reshaped[i,j], num_classes=num_letters)
    
    return new, output
```

```python
#define x and y
time_steps = 20
x, y = data_preprocess(data, time_steps, num_letters, char2int)
print(x.shape, len(y))
```

```python
print(x)
```

```python
#分离数据
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=10)
print(x_train.shape, len(y_train))    
```

```python
y_train_category = to_categorical(y_train, num_letters)
print(y_train_category)
```

```python
#建立模型
from os import access
from keras.models import Sequential
from keras.layers import LSTM, Dense
from torch import mode

model = Sequential()
model.add(LSTM(units=20, input_shape=(x_train.shape[1], x_train.shape[2]), activation='relu'))
#输出层 多分类
model.add(Dense(units=num_letters, activation='softmax'))
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
#训练模型
model.fit(x_train, y_train_category, epochs=5, batch_size=1000)
```

```python
#预测
y_train_predict = model.predict(x_train)
y_train_predict = np.round(y_train_predict).astype(int)
print(y_train_predict)
```

```python
#one hot 转换
labels = np.argmax(y_train_predict, axis=1)
print(labels)
y_train_predict_char = [int2char[i] for i in labels]
print(y_train_predict_char)
```

```python
from sklearn.metrics import accuracy_score
accuracy_train = accuracy_score(y_train, labels)
print("训练集准确率：", accuracy_train)
```

```python
#预测
y_test_predict = model.predict(x_test)
y_test_predict = np.round(y_test_predict).astype(int)
print(y_test_predict)
labels = np.argmax(y_test_predict, axis=1)
print(labels)
accuracy_test = accuracy_score(y_test, labels)
print("测试集准确率：", accuracy_test)
```

```python
new_letters = 'flare is a teacher in ai industry. He obtained his phd in Australia.'
x_new, y_new = data_preprocess(new_letters, time_steps, num_letters, char2int)
y_new_predict = model.predict(x_new)
y_new_predict = np.round(y_new_predict).astype(int)
labels = np.argmax(y_new_predict, axis=1)
y_new_predict_char = [int2char[i] for i in labels]
print(y_new_predict_char)
```
