```python
import pandas as pd

# 模拟数据：房屋面积 (平方米)、房间数、楼层、建造年份、位置（类别变量）
data = {
    'area': [70, 85, 100, 120, 60, 150, 200, 80, 95, 110],
    'rooms': [2, 3, 3, 4, 2, 5, 6, 3, 3, 4],
    'floor': [5, 2, 8, 10, 3, 15, 18, 7, 9, 11],
    'year_built': [2005, 2010, 2012, 2015, 2000, 2018, 2020, 2008, 2011, 2016],
    'location': ['Chaoyang', 'Haidian', 'Chaoyang', 'Dongcheng', 'Fengtai', 'Haidian', 'Chaoyang', 'Fengtai', 'Dongcheng', 'Haidian'],
    'price': [5000000, 6000000, 6500000, 7000000, 4500000, 10000000, 12000000, 5500000, 6200000, 7500000]  # 房价（目标变量）
}

data = pd.DataFrame(data)
print(data.head())
```

```python
from matplotlib import pyplot as plt
fig = plt.figure()
flg1 = plt.subplot(2,3,1)
plt.scatter(data['area'], data['price'])
plt.title('Area vs Price')

flg2 = plt.subplot(2,3,2)
plt.scatter(data['rooms'], data['price'])
plt.title('rooms vs Price')

flg3 = plt.subplot(2,3,3)
plt.scatter(data['floor'], data['price'])
plt.title('floor vs Price')

flg4 = plt.subplot(2,3,4)
plt.scatter(data['year_built'], data['price'])
plt.title('year_built vs Price')

flg5 = plt.subplot(2,3,5)
plt.scatter(data['location'], data['price'])
plt.title('location vs Price')

plt.show()
```

```python
x = data['area']
y = data['price']
from numpy import array
x = array(x).reshape(-1, 1)
y = array(y).reshape(-1, 1)

from sklearn.linear_model import LinearRegression
LR1 = LinearRegression()
LR1.fit(x, y)
y_predict = LR1.predict(x)
print(y_predict) 
print(y)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mse = mean_squared_error(y, y_predict)
r2 = r2_score(y, y_predict)

print("MSE: %.2f" % mse)
print("R2: %.2f" % r2)
```

```python
fig6 = plt.figure()
plt.scatter(x, y)
plt.plot(x, y_predict, color='red')
```

```python
df = pd.DataFrame(data)
x_multi = df[['area', 'rooms', 'floor', 'year_built']] 
y_multi = data['price'] 
```
