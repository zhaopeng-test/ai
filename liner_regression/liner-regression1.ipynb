```python
#pandas import data
import pandas as pd
data = pd.read_csv('xy_values.csv')

data.head()
```

```python
x = data.loc[:, 'x']
y = data.loc[:, 'y']

print(x, y)
```

```python
from matplotlib import pyplot as plt
plt.figure()
plt.scatter(x, y)
plt.show()
```

创建线性回归模型，通过数据进行训练

```python
from sklearn.linear_model import  LinearRegression

lr_mode = LinearRegression()

#print(x, y)
import numpy as np

x = np.array(x).reshape(-1, 1)
y = np.array(y).reshape(-1, 1)

print(type(x), x.shape)


lr_mode.fit(x, y)
```

用x 预测结果

```python
y_predict = lr_mode.predict(x)
print(y_predict)
print(x)
```

预测3.5 结果

```python
y_3 = lr_mode.predict([[3.5]])
print(y_3)
```

获取a, b 值

```python
a = lr_mode.coef_
b = lr_mode.intercept_

print(a, b)
```

评估模型
MSE 越小约好
R2 越接近1越好

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y, y_predict)
r2 = r2_score(y, y_predict)

print("MSE: %.2f" % mse)
print("R2: %.2f" % r2)
```
